```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Product_A=np.array([120,130,125])
Product_B=np.array([150,160,170])
Product_C=np.array([100,110,105])
Months=np.array(['Jan','Feb','Mar'])
data= {'Months':Months,
      'Product A':Product_A,
      'Product B':Product_B,
      'Product C':Product_C}
df=pd.DataFrame(data)
print(df)
print()
df.describe()
print()
plt.plot(Months,Product_A,label='Product A')
plt.plot(Months,Product_B,label='Product_B')
plt.plot(Months,Product_C,label='Product_C')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(loc='best',frameon=True)
plt.title('Monthly Sales Data for Three Products')
plt.tight_layout()
plt.show()
```
